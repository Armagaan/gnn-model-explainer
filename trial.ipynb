# Dataset Creation

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from utils import synthetic_structsim
from utils import featgen
from gengraph import perturb

%matplotlib inline
np.random.seed(1)
```

### BA shapes

```python
list_of_shapes = [['house']] * 4

basis, role_ids, shapes = synthetic_structsim.build_graph(
    width_basis=6,
    basis_type='ba',
    list_shapes=list_of_shapes
)

print(basis)
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.title('basis', fontsize=18)
nx.draw(basis, with_labels=True, font_weight='bold')

add_basis = perturb([basis], p=0.1, ptype='add')[0]
print(add_basis)
plt.subplot(2,2,2)
plt.title('add-perturb', fontsize=18)
nx.draw(add_basis, with_labels=True, font_weight='bold')

del_basis = perturb([basis], p=0.1, ptype='del')[0]
print(del_basis)
plt.subplot(2,2,3)
plt.title('del-perturb', fontsize=18)
nx.draw(del_basis, with_labels=True, font_weight='bold')

del_basis = perturb([basis], p=0.1, ptype='mix')[0]
print(del_basis)
plt.subplot(2,2,4)
plt.title('mix-perturb', fontsize=18)
nx.draw(del_basis, with_labels=True, font_weight='bold')
plt.show()
```
